```python
from transformers import pipeline

generator = pipeline('text-generation', model='gpt2')
```

```python
results = generator("artificial intelligence is", max_length=300)
results[0]['generated_text']
```

```python
results = generator("gpt-2 model is", max_length=50)
results[0]['generated_text']
```
