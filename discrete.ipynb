```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-2, 2, 100)  # Adjust the range and number of points as needed

# Calculate y values
y = np.exp(x)

# Create the plot
plt.plot(x, y, label='$e^x$')
plt.axhline(0, color='k', linewidth=0.8)  # Add a dark horizontal line at y=0
plt.axvline(0, color='k', linewidth=0.8)  # Add a dark vertical line at x=0

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = e^x$')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
#programmed by Khagendra Khatri
```
